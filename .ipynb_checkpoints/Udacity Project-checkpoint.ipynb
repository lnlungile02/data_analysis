### 3. Look at which genre combination gave the most revenue?

```python
# Use this, and more code cells, to explore your data. Don't forget to add
#   Markdown cells to document your observations and findings.
#Look at which genre gave the most revenue?
df_movies.genres.unique()
```

```python
# Create a Container Dictionary
averages = {}

# Split the data into different regions
for genre in df_movies['genres'].unique():
    df_genres = df_movies[df_movies['genres'] == genre]

    # Apply an aggregation function
    average = df_genres['revenue'].mean()

    # Combine the data into a DataFrame
    averages[genre] = [average]

aggregate_df = pd.DataFrame.from_dict(averages, orient='index', columns=['Average_Revenue'])
aggregate_df.head()
```

```python
aggregate_df = aggregate_df[aggregate_df.Average_Revenue > 0]
aggregate_df=aggregate_df.sort_values("Average_Revenue",ascending=False)
aggregate_df=aggregate_df.head(20)
aggregate_df.head()
```

```python
aggregate_df.plot.barh()
plt.xlabel("Genre Combinations")
plt.ylabel("Average Revenue")
plt.title("Graph showing Genre with the highest Average Revenue")
plt.grid(True)
plt.show()
```

<h5> Observation:</h5> The Combination of Crime|Drama|Mystery|Thriller|Action is the highest earning genre combination

### 5. What is the relationship between budget and revenue?

```python
def scatter_chart(column1,column2,x_label,y_label,graph_heading):
    
    x=df_movies[column1]
    y=df_movies[column2]
    plt.scatter(x,y)
    #calculate equation for trendline
    z = np.polyfit(x, y, 1)
    p = np.poly1d(z)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(graph_heading)
    plt.grid(True)
    #add trendline to plot
    plt.plot(x, p(x),color="red")
```

```python
scatter_chart("budget","revenue","budget","revenue","Graph showing Relationship between Budget vs Revenue")
```

<h5>Observation:</h5> From the graph we see that the scatterplot is quite close. There seems to be a positive relationship between the variables. This shows that here is a some correlation between budget and revenue.

```python
x=popularity_df['%_Popularity']
y=aggregate_df['Average_Revenue']

plt.scatter(x,y)

#calculate equation for trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.xlabel('%_Popularity')
plt.ylabel('Average_Revenue')
plt.title("Graph Showing Popularity versus Average Revenue")
plt.grid(True)
#add trendline to plot
plt.plot(x, p(x),color="red")
```

### 13. Which director has the highest earning movies?

```python
#calculate the total revenue for the films directed by the director
def barh_plot(column1,column2,sort_column,drop_unknown,title,x_label,y_label):
    director_revenue = df_movies.groupby(column1)[column2].sum()
    director_revenue=pd.DataFrame(director_revenue)
    director_revenue=director_revenue.sort_values(sort_column,ascending=False)
    director_revenue.drop(director_revenue[director_revenue[drop_unknown]==0].index,inplace=True)
    director_revenue.head(20).plot.barh()
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid()
```

```python
barh_plot('director','revenue',"revenue","revenue","Top 20 Directors who generated the Highet Revenue","Revenue","Director")
```

<h5>Observation:</h5> From the graph we see the top 20 directors with the highest revenue . Steven Speilberg has the highest revenuet and RSam Mendes the least.

### 15. Which production company had the highest revenue?

```python
#calculate the revenue of each production company
produced_movies_revenue = df_movies.groupby('production_companies')['revenue'].sum()
produced_movies_revenue=pd.DataFrame(produced_movies_revenue)
produced_movies_revenue=produced_movies_revenue.sort_values("revenue",ascending=False)
produced_movies_revenue.head(20).plot.barh()
plt.title("Top 20 Production Company Revenue")
plt.xlabel("Revenue")
plt.ylabel("Production Comapnies")
plt.grid(True)
```

<h5>Observation:</h5> From the graph we see the top 20 Production companies with the most revenue. Paramount Pictures has the highest revenue and Lionsgate|Color Force earned the least.

```python
prod_co_releases = df_movies.groupby('release_year')['production_companies'].count().plot(color="red")
plt.title("Production companies that released moves over the years")
plt.xlabel("Release year")
plt.ylabel("No of Production Comapnies")
plt.grid(True)
```
